```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
print(" Summary Statistics ")
print(df.describe())
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
df[numeric_cols].hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()
```

```python
sns.pairplot(df[numeric_cols])
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
print("\n=== Pattern & Trend Observations ===")
# Example: Survival rate by gender
print(df.groupby('Sex')['Survived'].mean())
# Example: Survival rate by passenger class
print(df.groupby('Pclass')['Survived'].mean())
```

```python
print("\n=== Basic Inferences ===")
print("1. Females had higher survival rate than males.")
print("2. Higher passenger class (1st) had better survival rates.")
print("3. Some extreme Fare values indicate wealthy passengers (possible outliers).")
print("4. Younger passengers had slightly higher survival chances.")
```
